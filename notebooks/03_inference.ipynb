```python
import os
import torch
from unsloth import FastLanguageModel

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

model_path = "unsloth/Qwen3.5-2B"
lora_path  = "/content/drive/MyDrive/svg_qwen_lora_v2/best_adapter"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_path,
    max_seq_length = 2048,
    load_in_4bit = True,
    dtype = None,
    device_map = {"": 0},
    fix_tokenizer = True,
)

model.load_adapter(lora_path)
model.to("cuda")
model.eval()
FastLanguageModel.for_inference(model)

print("✅ Model & LoRA fully synced on GPU 0. Ready.")
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import re
from tqdm import tqdm

ALLOWED_TAGS = {
    "svg","g","path","rect","circle","ellipse","line","polyline","polygon",
    "defs","use","symbol","clipPath","mask","linearGradient","radialGradient",
    "stop","text","tspan","title","desc","style","pattern","marker","filter"
}

FALLBACK_SVG = (
    '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 256 256" width="256" height="256">'
    '<rect x="0" y="0" width="256" height="256" fill="white"/>'
    '<circle cx="128" cy="128" r="64" fill="black"/>'
    '</svg>'
)

def clean_svg_markup(text):
    if "</think>" in text:
      text = text[text.find("</think>") + len("</think>"):]
    start = text.find("<svg")
    if start == -1:
        return FALLBACK_SVG
    end = text.find("</svg>")
    svg_code = text[start:end+6] if end != -1 else text[start:]

    if len(svg_code) > 8000:
        svg_code = svg_code[:7993]

    if len(re.findall(r'<path', svg_code, re.IGNORECASE)) > 256:
        return FALLBACK_SVG

    try:
        root = ET.fromstring(svg_code)
    except ET.ParseError:
        try:
            root = ET.fromstring(svg_code + "</svg>")
        except ET.ParseError:
            last_complete = svg_code.rfind("/>")
            last_open_close = svg_code.rfind(">")
            cut = max(last_complete, last_open_close)
            if cut != -1:
                svg_code = svg_code[:cut+1] + "</svg>"
                try:
                    root = ET.fromstring(svg_code)
                except ET.ParseError:
                    return FALLBACK_SVG
            else:
                return FALLBACK_SVG

    for elem in root.iter():
        tag = elem.tag.split("}")[-1] if "}" in elem.tag else elem.tag
        if tag not in ALLOWED_TAGS:
            return FALLBACK_SVG

    root.set("viewBox", "0 0 256 256")
    root.set("width", "256")
    root.set("height", "256")
    # root.set("xmlns", "http://www.w3.org/2000/svg")

    ET.register_namespace('', 'http://www.w3.org/2000/svg')
    return ET.tostring(root, encoding="unicode")

test_path = "data/test.csv"
test_df = pd.read_csv(test_path)
print(f"✅ Loaded {len(test_df)} test rows")
```

```python

SYSTEM_PROMPT = """Output SVG code only. No explanations.
Format: <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 256 256" width="256" height="256">SHAPES</svg>
Use only: rect, circle, ellipse, line, polygon for shapes.
Maximum 8 shapes. Keep total under 500 characters.
Always end with </svg>."""

row = test_df.iloc[0]
messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": f'Generate SVG for: {row["prompt"]}'},
]
input_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
inputs = tokenizer(text=input_text, return_tensors="pt").to("cuda")

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=800,
        temperature=0.7,
        do_sample=True,
        repetition_penalty=1.05,
        pad_token_id=tokenizer.eos_token_id,
    )

raw = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True).strip()
print("RAW:")
print(raw[:400])
print("\nAFTER CLEAN:")
print(clean_svg_markup(raw)[:400])
```

```python
import os

# Delete old submission
if os.path.exists("submission.csv"):
    os.remove("submission.csv")
    print("🗑️ Deleted old submission. Starting fresh.")

SYSTEM_PROMPT = """Output SVG code only. No explanations.
Format: <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 256 256" width="256" height="256">SHAPES</svg>
Use only: rect, circle, ellipse, line, polygon for shapes.
Maximum 8 shapes. Keep total under 500 characters.
Always end with </svg>."""

submission_path = "submission.csv"
if os.path.exists(submission_path):
    recovered_df = pd.read_csv(submission_path)
    results = recovered_df.to_dict('records')
    start_index = len(results)
    print(f"✅ Recovered {start_index} rows! Resuming from row {start_index}...")
else:
    results = []
    start_index = 0
    print("⚠️ No existing submission found. Starting from scratch.")

for i, row in tqdm(test_df.iloc[start_index:].iterrows(), total=len(test_df)-start_index):
    try:
        messages = [
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": f'Generate SVG for: {row["prompt"]}'},
        ]

        input_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        inputs = tokenizer(text=input_text, return_tensors="pt").to("cuda")

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=400, # from 800
                repetition_penalty=1.05,
                use_cache=True,
                pad_token_id=tokenizer.eos_token_id,
                temperature=0.7,
                do_sample=True
            )

        raw_output = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True).strip()
        svg_output = clean_svg_markup(raw_output)
        results.append({"id": row['id'], "svg": svg_output})

        if len(results) <= 3:
            print(f"Row {len(results)} preview: {results[-1]['svg'][:200]}")

    except Exception as e:
        results.append({"id": row['id'], "svg": FALLBACK_SVG})

    if len(results) % 10 == 0:
        pd.DataFrame(results).to_csv("submission.csv", index=False)
        pd.DataFrame(results).to_csv("/content/drive/MyDrive/submission_backup.csv", index=False)
        print(f"💾 Auto-saved {len(results)} rows total.")

pd.DataFrame(results).to_csv("submission.csv", index=False)
print(f"🏁 Done! Final file has {len(results)} rows.")

# Fill missing rows with fallback
done_ids = {r['id'] for r in results}
for _, row in test_df.iterrows():
    if row['id'] not in done_ids:
        results.append({"id": row['id'], "svg": FALLBACK_SVG})

pd.DataFrame(results).to_csv("submission.csv", index=False)
print(f"✅ Final file with fallbacks: {len(results)} rows.")
```
